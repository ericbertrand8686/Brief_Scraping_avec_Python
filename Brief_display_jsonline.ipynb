```python
import jsonlines
import numpy as np
import pandas as pd
```

```python
def default_value(x):
   return (x if x is not None else "None")
```

```python
muzeo_data = []

with jsonlines.open("brief_muzeo_scraping/muzeo_data.jsonlines") as reader:
    for obj in reader:
        muzeo_data.append(obj)
```

```python
titre = []
idproduit = []
prix = []
delfab = []
urlim = []

for i, manga in enumerate(muzeo_data) :
    titre.append(default_value(manga['Titre']))
    idproduit.append(default_value(manga['ID produit']))
    prix.append(default_value(manga['Prix']))
    delfab.append(default_value(manga['Delai de fabrication']))
    urlim.append(default_value(manga["Url de l'image"]))
    
    df = pd.DataFrame({'Titre' : titre,'ID produit' : idproduit, 'Prix' : prix, 'Delai de fabrication' : delfab, "Url de l'image" : urlim})
```
